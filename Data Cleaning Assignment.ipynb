**Q2. Handling Missing Age
Scenario:Age is critical for medical analysis, but some values are missing.
Task:
Replace missing Age values with an appropriate method
Justify your choice (mean/median)**

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df
```

**Missing Age values are handled using the median imputation method. The median is preferred over the mean because it is robust to outliers and provides a more accurate central value for skewed data. This ensures that the dataset remains consistent and suitable for analysis.**

**Q3. Handling Missing Treatment Cost
Scenario:
Treatment cost is highly skewed due to expensive treatments.
Task:
Handle missing Treatment_Cost values
Choose the correct imputation method and explain why**

```python
df['Treatment_Cost']=df['Treatment_Cost'].fillna(df['Treatment_Cost'].median())
df
```

**Why Median is Used:**
- Median is used because the dataset is highly skewed.
- Median is the middle value.
- It is not affected by extreme values (outliers).
- It gives a more realistic replacement.
- Maintains the overall distribution of the data.
- Prevents bias caused by extreme values.
- Keeps the dataset balanced and reliable.
- Improves the accuracy of analysis and models.

**Q6. Outlier Detection (Treatment Cost)
Scenario:
Extreme treatment costs are affecting analysis.
Task:
Detect outliers using IQR method
Display number of outliers**

```python
Q1 = df['Treatment_Cost'].quantile(0.25)
Q3 = df['Treatment_Cost'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df['Treatment_Cost'] < lower) | (df['Treatment_Cost'] > upper)]
print(outliers)
```

**Q7. Outlier Treatment
Scenario:
The business team wants to retain all records.
Task:
Apply capping (Winsorization) on Treatment_Cost
Use 5th and 95th percentile**

```python
lower_cap = df['Treatment_Cost'].quantile(0.05)
upper_cap = df['Treatment_Cost'].quantile(0.95)
df['Treatment_Cost'] = df['Treatment_Cost'].clip(lower_cap, upper_cap)
df['Treatment_Cost']
```

**Q9. Time-Based Missing Handling
Scenario:
Admission dates should follow a logical sequence.
Task:
Sort data by Admission_Date
Apply forward fill or backward fill where appropriate
Justify your choice**

```python
df['Admission_Date'] = pd.to_datetime(df['Admission_Date'])
df = df.sort_values(by='Admission_Date')
df = df.fillna(method='ffill')
df
```
